## `3- .loc Attribute`

```python
songs3=pd.Series([145,142,38,13,2],name='counts',index=['Paul','John','George','Ringo','George'])
```

### 3-1 scalar value of one of the index labels


```python
songs3.loc['Paul']
```

```python
songs3.loc['George']
```

### 3-2 list of index labels.


```python
city2.loc[['Ferrari','Subaru']]
```

### 3-3 slice of labels (closed interval so it includes the stop value)


```python
# slicing with .loc, it follows the closed interval. The closed interval includes both the start index and the final index.
# this will throw error.
songs3=pd.Series([145,142,38,13,2],name='counts',index=['Paul','John','George','Ringo','George'])
songs3.sort_index().loc['George':'Paul']
```

```python
# Note that slicing with labels is closed and includes the end value.
songs3.sort_index()
```

```python
songs3.sort_index().loc['G':'J']
```

### 3-4 index



```python
#pd.Index is a class that represents an immutable array-like structure, specifically used for indexing and axis labels
songs3.loc[pd.Index(['George','George'])]
```

## `4 .iloc Attribute`

#### 4-1 scalar index position (an integer)


```python
songs3
```

```python
songs3.iloc[2]
```

#### 2- list of index positions

```python
# return a series object
songs3.iloc[[0,3]]
songs3.iloc[[0,-5]]
```

#### 3- slice of positions (half-open interval so it does not include stop value).


```python
songs3.iloc[0:2]
songs3.iloc[-2:]
```

#### 4- NumPy array (or Python list) of boolean values.

```python
mask=city2>50
```

```python
mask.to_numpy()
```

```python
city2.iloc[mask.to_numpy()]
```

## `7- Filtering Index Values`
Note that exact match (with items) fails with duplicate index labels.
items, like, and regex parameters,

```python
city2.filter(like='MINI')
```

```python
city2.filter(items=['Chevrolet','Jeep'])
```

```python
city2.filter(regex='(Ford)|(Subaru)')
```

#### `8- Reindex`
The .reindex() method in Python, particularly in pandas, is used to conform a DataFrame or Series to a new index. It returns a new object with the data aligned to a new set of labels.

```python
# Create a simple Series
data = {'A': 10, 'B': 20, 'C': 30}
original_series = pd.Series(data)

# Create a new index
new_index = ['A', 'B', 'D']

# Use reindex to align the Series with the new index
new_series = original_series.reindex(new_index)

print(new_series)
```
